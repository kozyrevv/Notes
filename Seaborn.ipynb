## Matrix Plots 

```python
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
## implement correlation between columns
tips.corr()
```

## Heatmaps 

```python
## aplying heatmap on datasets
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(), cmap='coolwarm', annot=True)
```

```python
flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
pvflights=flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(pvflights)
```

```python
sns.heatmap(pvflights, cmap='magma', linecolor='white', linewidths='1')
```

```python
sns.heatmap(pvflights, cmap='coolwarm', linecolor='white', linewidths='1')
```

## Clustermaps 

```python
sns.heatmap(pvflights)
```

```python
sns.clustermap(pvflights)
```

```python
sns.clustermap(pvflights, cmap='coolwarm')
```

```python
sns.clustermap(pvflights, cmap='coolwarm', standard_scale=1)
```

## Grids 

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
