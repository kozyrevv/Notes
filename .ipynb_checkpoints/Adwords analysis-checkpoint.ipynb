```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 
```

```python
report=pd.read_excel('sem.xlsx')
df=pd.DataFrame(report)
```

```python
df.head()
```

```python
device_sem=df[(df['Quality score']>0)& (df['Clicks'])&(df['Device'])]
```

```python
sns.distplot(device_sem['Quality score'])
```

```python
sns.pairplot(device_sem, hue='Device')
```

```python
sns.heatmap(device_sem.corr())
```
