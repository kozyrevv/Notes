```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
# Functional method of visualization with matplotlib 
# plt.show() to print plot without jupyter notebook

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot()
```

```python
# Object Oriented Method

fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
# Creates blank canvas
fig = plt.figure()

axes0 = fig.add_axes([0.2,0.2,1.1,1.1]) # big axes
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Big Test Axes 0
axes0.set_title('BIG AXES')

# Larger Figure Axes 1
axes1.plot(x, y, 'o')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');

```

```python
### LABELS ATTACHING + LEGEND CHECKING AND LOCATION

fig=plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3, label= 'X Cubed')

ax.legend(loc=(0.1,0.2))

plt.show()
```

### Training time 

```python
### single Plot. Labels, legend.location 
graph = plt.figure()
ax= graph.add_axes([0,0,1,1])


ax.plot(y,y**2,label='Y Squared')
ax.plot(y,y**3,label='Y Cubed')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Single Graph')

plt.legend(loc=(0.1,0.2))

plt.show()

```

```python
### Object oriented method ! multiple subplots, indexing, dimensions.

subgraph,axes=plt.subplots(figsize=(8,4), nrows=1, ncols=3)

axes[0].plot(x,y,'o')
axes[1].plot(x,x,'b')
axes[2].plot(y,y,'g')

axes[0].set_title('XY Multiple Subplots')


plt.tight_layout()


```

# Plot Appearance 

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF7C00',linewidth=5,alpha=0.5) 

plt.show()
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF7C00',linewidth=5,linestyle='steps') ## lw=line width, ls = line style 

plt.show()
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',alpha=0.5,lw=2, ls='-', marker='o', markersize=20,
       markerfacecolor='yellow',markeredgewidth=2, markeredgecolor='blue') 

plt.show()
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',alpha=0.5,lw=2, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])

plt.show()
```
